**Check if the matrix is diagonal** 

```python
def isDiag(M): 
    i, j = np.nonzero(M)
    return np.all(i == j)

#img is too big. 

matrix1 = np.array ([[1,2,3], [2,6,8], [7,9,9]])
matrix2 = np.array ([[1,0,0], [0,6,0], [0,0,9]])

print (matrix1)
print (isDiag(matrix1))
print (matrix2)
print (isDiag(matrix2))
```

**Norm1**

```python
from numpy.linalg import norm
l1=norm (matrix1,1)
print (l1)
```

**Norm2**

```python
l2 = norm(matrix1)
print (l2)
```
